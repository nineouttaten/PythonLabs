### Бинарный поиск

#### В упорядоченном массиве находится средний элемент. Дальше 3 варианта развития событий:
#### 1. Если средний элемент равен искомому, поиск заканчивается
#### 2. Если средний элемент больше искомого, то правая граница сдвигается к среднему и опять находится средний
#### 3. Если средний элемент меньше искомого, то левая граница сдвигается к среднему и опять находится средний
#### Повторяется пока не найдет

```python
def binary_search(mass, sought):
    left = 0
    right = len(mass) - 1
    mid = (left + right) // 2
    while mass[mid] != sought and left <= right:
        if sought > mass[mid]:
            left = mid + 1
        else:
            right = mid - 1
        mid = (left + right) // 2

    if left > right :
        return "Нету"
    else:
        return mid
        
```

```python
sought = int(input("Введите искомое число "))
start_time = time.time()
test(binary_search(big_mass, sought), binary_search(lil_mass, sought))
```

### Фибоначчиев поиск
#### Элемент ищется по индексам равным числам фибоначчи 
#### Если очередной элемент больше искомого, то массив обрезается так, что левая граница это прошлое число фибоначчи, а правая это последнее взятое. Повторять пока очередной элемент не будет равен искомому

```python
def fibonacci_search(mass, sought):
    obrez = 0
    past = 1
    present = 1
    while mass[past] != sought:
        past = 1
        present = 1
        while mass[present - 1] < sought:
            bubble = present
            present += past
            past = bubble
            if present > len(mass):
                present = len(mass)
        mass = mass[:present]
        mass = mass[past:]
        obrez += past 
        present = len(mass)
        past = 0
    return obrez
```

```python
sought = int(input("Введите искомое число "))
start_time = time.time()
big_mass_copy = big_mass.copy()
lil_mass_copy = lil_mass.copy()
test(fibonacci_search(big_mass_copy, sought), fibonacci_search(lil_mass_copy, sought))
```

#### Фибоначчиев поиск ищет быстрее в маленьких массивах, а стандартная функция быстрее в больших

### Интерполяционный поиск
#### Схож с бинарным поиском, но средний элемент высчитывается не средним значением, а специальной формулой

```python
def interpolation_search(mass, sought):
    left = 0
    right = (len(mass) - 1)
    while left <= right and sought >= mass[left] and sought <= mass[right]:
        index = left + int(((float(right - left) / ( mass[right] - mass[left])) * ( sought - mass[left])))
        if mass[index] == sought:
            return index
        if mass[index] < sought:
            left = index + 1
        else:
            right = index - 1
```

```python
sought = int(input("Введите искомое число "))
start_time = time.time()
big_mass_copy = big_mass.copy()
lil_mass_copy = lil_mass.copy()
test(interpolation_search(big_mass_copy, sought), interpolation_search(lil_mass_copy, sought))
```

### Бинарное дерево
#### Строится бинарное дерево, в котором у каждого элемента может быть две ветки: левая с объектом меньше и правая с объектом больше. Поиск по этому дереву ведется по принципу - если искомое число меньше, чем сравниваемое число в дереве, то сравнение продолжается по левой ветке сравниваемого числа, если больше, то по правой. Перебирание веток происходит пока не найдется искомое число

```python
class Tree:

    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data

    def insert(self, data):
        if self.data:
            if data < self.data:
                if self.left is None:
                    self.left = Tree(data)
                else:
                    self.left.insert(data)
            elif data > self.data:
                if self.right is None:
                    self.right = Tree(data)
                else:
                    self.right.insert(data)
        else:
            self.data = data

    def findNum(self, lkpval):
        if lkpval < self.data:
            if self.left is None:
                return str(lkpval)+" Not Found"
            return self.left.findNum(lkpval)
        elif lkpval > self.data:
            if self.right is None:
                return str(lkpval)+" Not Found"
            return self.right.findNum(lkpval)
        else:
            print(str(self.data) + ' is found')

    def PrintTree(self):
        if self.left:
            self.left.PrintTree()
        print( self.data),
        if self.right:
            self.right.PrintTree()
```

#### Создание дерева по массиву

```python
tree = Tree(tree_mass[0])
for i in range(1, len(tree_mass)):
    tree.insert(tree_mass[i])
```

#### Поиск по дереву

```python
start_time = time.time()
tree.findNum(3456)
print(time.time() - start_time)
```

## Задача 2
#### Функция хеширования 

```python
def hash(number, size):
    number += 10000
    size = size // 20000
    return number * size
```

#### Массив для теста рехеширования

```python
hash_mass = []
for i in range(30000):
    hash_mass.append(randint(-10000, 10000))
```

### Простое рехеширование
#### Если место для очередного элемента занято, то рассматривается следующее место, затем еще следующее и т.д.

```python
start_time = time.time()
rehashed = [None] *len(hash_mass)
size = len(hash_mass)
for j in range(len(hash_mass)):
    i = 0
    if rehashed[hash(hash_mass[j],size)] == None:
        rehashed[hash(hash_mass[j], size)] = hash_mass[j]
    else:
        while rehashed[hash(hash_mass[j], size) + i % size] != None:
            i += 1
        rehashed[hash(hash_mass[j], size) + i % size] = hash_mass[j]
simpleHashTime = time.time() - start_time
print(simpleHashTime)
```

### Рехеширование с помощью псевдослучайных чисел
#### Если место для очередного элемента занято, для него выбирается псевдослучайное место, до тех пор, пока не найдется подходящее.

```python
start_time = time.time()
def randHash(number, size):
    return hash(number, size)
rehashed = [None] *len(hash_mass)
size = len(hash_mass)
for j in range(len(hash_mass)):
    i = 0
    if rehashed[randHash(hash_mass[j],size)] == None:
        rehashed[randHash(hash_mass[j], size)] = hash_mass[j]
    else:
        while rehashed[i] != None:
            i = randint(1, size - 1)
        rehashed[i] = hash_mass[j]
randomHashTime = time.time() - start_time
print(randomHashTime)
```

### Рехеширование цепочкой
#### Если место для очередного элемента занято, создается цепочка, в которую будут добавлятся все элементы с таким же кодом

```python
start_time = time.time()
rehashed = [None] * len(hash_mass)
for i in range(len(hash_mass)):
    if rehashed[hash(hash_mass[i], len(hash_mass))] == None:
        rehashed[hash(hash_mass[i], len(hash_mass))] = hash_mass[i]
    elif rehashed[hash(hash_mass[i], len(hash_mass))] != None:
        chain = rehashed[hash(hash_mass[i], len(hash_mass))]
        if type(chain) != "list":
            rehashed[hash(hash_mass[i], len(hash_mass))] = []
            if type(chain) is list:
                for j in range(len(chain)):
                    rehashed[hash(hash_mass[i], len(hash_mass))].append(chain[j])
            else:
                rehashed[hash(hash_mass[i], len(hash_mass))].append(chain)
        rehashed[hash(hash_mass[i], len(hash_mass))].append(hash_mass[i])
chainHashTime = time.time() - start_time
print(chainHashTime)
```

### Сравнение всех результатов

```python
print("Простое рехеширование", simpleHashTime)
print("Рехеширование рандомом", randomHashTime)
print("Рехеширование цепочками", chainHashTime)
```
